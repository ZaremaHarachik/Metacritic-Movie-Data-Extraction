# Metacritic Movie Scraper
This notebook fetches movie data from Metacritic and extracts information such as titles and descriptions using regular expressions.

```python
# Import required libraries for web scraping and data processing

# urllib3 is used to send HTTP requests and fetch webpage content
import urllib3

# certifi provides SSL certificates to securely access HTTPS websites
import certifi

# re is Python’s built-in library for working with regular expressions
# used to search and extract patterns from HTML text
import re

# pandas is used to organize extracted data into tables (Data frames)
import pandas as pd
```

```python
# Create an HTTP connection pool
# This allows to send requests to websites
http = urllib3.PoolManager(ca_certs=certifi.where())

# Metacritic URL for movies (year 2020)
url = "https://www.metacritic.com/browse/movie/all/all/2020/metascore/"

# Headers to simulate a real browser
# Some websites block requests without a User-Agent
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Send a GET request to the website
response = http.request('GET', url, headers=headers)

# Decode the response from bytes to readable text (UTF-8)
# Decoding is necessary because web responses are returned as bytes.
html = response.data.decode('utf-8')

# Print length of HTML to confirm the page was loaded successfully
print("HTML length:", len(html))
```

```python
# Extract movie titles using regex
# This pattern looks for text inside data-title="..."
titles = re.findall(r'data-title="(.*?)">', html)

# Print how many titles were found to verify extraction
print("Titles found:", len(titles))
# print all 24 titles
print(titles)
```

```python
# Extract descriptions using regex
# This pattern finds text inside <span> tags related to descriptions
descriptions = re.findall(r'0"><span>([\s\S]*?)<\/span>', html)

# Clean HTML encoding
# Replace &#39; with a normal apostrophe '
descriptions = [d.replace("&#39;", "'") for d in descriptions]

# Filter results to keep only real descriptions
# Short matches (like numbers or small text) are removed
descriptions = [d for d in descriptions if len(d) > 100]

# Print results to verify correct number of descriptions
print("Descriptions found:", len(descriptions))
```

```python
# Extract release dates using regex
# The regex looks for a capitalized month (3 letters), followed by day and year.
# This pattern matches dates like "Nov 20, 2020"
dates = re.findall(r'[A-Z][a-z]{2} \d{1,2}, \d{4}', html)

# Print results to check how many dates were found
print("Dates found:", len(dates))
print(dates)
```

```python
# Extract metascores using regex
# The regex looks for the exact phrase "Metascore" and captures the number.
# helps avoid matching random numbers in the HTML.
# This pattern finds numbers inside title="Metascore XX out of 100"
scores = re.findall(r'title="Metascore (\d+) out of 100"', html)

# Print results to verify correct number of scores
print("Scores found:", len(scores))
print(scores)

```

```python
# Create an empty list to store all movie data
movies = []

# Loop through all lists at the same time
# use min( ) to avoid index errors if lists have different lengths
for i in range(min(len(titles), len(descriptions), len(dates), len(scores))):

    # For each index, combine data into a dictionary
    # This keeps title, description, date, and metascore together for each movie
    movies.append({
        "Title": titles[i],              # movie title
        "Description": descriptions[i],  # movie description
        "Date release": dates[i],        # release date
        "Metascore": scores[i]           # metascore rating
    })

# This step converts separate lists into a structured format.
# Each movie is stored as one dictionary, which makes it easier to analyze and convert into a DataFrame later.
    
```

```python
# Convert the list of dictionaries into a pandas DataFrame
# This organizes the data into a table format (rows and columns)
df = pd.DataFrame(movies)

# Display the DataFrame to check the final result
df
```
